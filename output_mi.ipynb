```python
%load_ext autoreload
%autoreload 2

import numpy  as np
import pandas as pd

from matplotlib import pyplot as plt

from common import hallmark as hm
```

```python
def calmi(path, hr=10, N=64):
    df = pd.read_csv(path, sep='\t')

    m  = df.Fsum <= 2.4 * 1000    
    w  = np.where(~m)[0]
    if len(w) != 0:
        print(f'WARNING: bring frame found: {w}; SKIP that frame')
        
    df = df[m]
    dt = (df.time_hr.iloc[-1] - df.time_hr.iloc[0] - hr) / N
    mi = []
    for i in range(N):
        t0  = df.time_hr[0] + i * dt
        sel = df[(t0 <= df.time_hr) & (df.time_hr < t0 + hr)]
        mi.append(np.std(sel.Fsum) / np.mean(sel.Fsum))
    return np.array(mi)
```
